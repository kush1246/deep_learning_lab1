```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
X=np.array([
    [0,0],[0,1],[1,0],[1,1]
])

y=np.array([[0],[1],[1],[0]])
```

```python
input_size=X.shape[1]
hidden_size=2
output_size=1

W1=np.random.randn(input_size, hidden_size)*0.01
b1=np.zeros((1, hidden_size))

W2=np.random.randn(hidden_size,output_size)*0.01
b2 = np.zeros((1, output_size))
```

```python
z1 = np.dot(X, W1) + b1

a1 = sigmoid(z1)

z2 = np.dot(a1, W2) + b2

y_hat = sigmoid(z2)
```

```python
error_output = y_hat - y
delta_output = error_output * sigmoid_derivative(z2)
error_hidden = np.dot(delta_output, W2.T)
error_hidden = np.dot(delta_output, W2.T)
delta_hidden = error_hidden * sigmoid_derivative(z1)
```

```python
learning_rate = 0.1
W2 -= learning_rate * np.dot(a1.T, delta_output)
b2 -= learning_rate * np.sum(delta_output, axis=0, keepdims=True)
W1 -= learning_rate * np.dot(X.T, delta_hidden)
b1 -= learning_rate * np.sum(delta_hidden, axis=0, keepdims=True)
```

```python
epochs = 1000
losses = []

for epoch in range(epochs):

    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, W2) + b2
    y_hat = sigmoid(z2)

    loss = np.mean((y_hat - y) ** 2)
    losses.append(loss)

    error_output = y_hat - y
    delta_output = error_output * sigmoid_derivative(z2)

    error_hidden = np.dot(delta_output, W2.T)
    delta_hidden = error_hidden * sigmoid_derivative(z1)

    W2 -= learning_rate * np.dot(a1.T, delta_output)
    b2 -= learning_rate * np.sum(delta_output, axis=0, keepdims=True)

    W1 -= learning_rate * np.dot(X.T, delta_hidden)
    b1 -= learning_rate * np.sum(delta_hidden, axis=0, keepdims=True)
```

```python
import matplotlib.pyplot as plt

plt.plot(losses)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss vs Epochs")
plt.show()
```
